```python
import csv
import pandas as pd
```

```python
df_main = pd.read_csv('hpo_xref_data.txt', delimiter = "\t")
df_main.head()
```

```python
df_UMLS = pd.read_csv('UMLS-DATA.txt', delimiter = "\t")
df_UMLS.head()
```

```python
for idx in range(len(df_main['DATABASE-NAME'])):
    if df_main.iloc[idx]['DATABASE-NAME'] == "UMLS":
        target = df_main.iloc[idx]['EXTERNAL-ID']
        target = df_UMLS[df_UMLS['CUI'].isin([target])]['STR']
        targetSet = set()
        for item in target:
            targetSet.add(item)
#         print(targetSet)
        if len(targetSet) > 0:
            df_main.iloc[idx]['EXTERNAL-NAME'] = targetSet
```

```python
df_main.to_csv(r'hpo_xref_data.txt', index=None, sep='\t')
```
