```python
import pandas as pd
```

```python
df_main = pd.read_csv('hpo_xref_data.txt', delimiter = "\t")
df_main.head()
```

```python
df_SNO = pd.read_csv('SnomedCT.txt', delimiter = "\t")
df_SNO.head()
```

```python
for idx in range(len(df_main['DATABASE-NAME'])):
    if df_main.iloc[idx]['DATABASE-NAME'] == "SNOMEDCT_US":
        target = df_main.iloc[idx]['EXTERNAL-ID']
        target = df_SNO[df_SNO['conceptId'].isin([target])]['term']
        if len(target) > 0:
#             print(target.iloc[0])
            df_main.iloc[idx]['EXTERNAL-NAME'] = (target.iloc[0])
```

```python
df_main.to_csv(r'hpo_xref_data.txt', index=None, sep='\t')
```
